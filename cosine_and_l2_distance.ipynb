### Functions

````python
import re

def extract_strings_gly(filename):
  if filename.endswith(".jpg"):
    # Split the filename into parts
    parts = filename.rsplit("_", 2)
    if len(parts) == 3:
      new_name = f"{parts[0]}_{parts[2]}"
      return new_name
  return filename

def get_avg_csv(input_path, output_path):
  df = pd.read_csv(input_path, header=None)
  df_sorted = df.sort_values(by=0, ascending=True)
  df_sorted['image'] = df_sorted[0]
  df_sorted['base_image'] = df_sorted['image'].apply(extract_strings_gly)
  df_avg_df = df_sorted.groupby('base_image').mean(numeric_only=True).reset_index()
  df_avg_df.to_csv(output_path, index=False)
  return df_avg_df
````

## Cosine and L2 distance

````python
def calculate_distances(vec1, vec2):
    l2_dist = euclidean(vec1, vec2)
    cosine_dist = cosine(vec1, vec2)
    return l2_dist, cosine_dist

# Function to retain top-k values in each vector
def retain_top_k(vec, k):
    vec_copy = vec.copy()
    threshold = np.partition(vec_copy, -k)[-k]  # Find the k-th largest value
    vec_copy[vec_copy < threshold] = 0  # Zero out all values below the threshold
    return vec_copy

def get_cosine_and_l2_distance(csv1_path, csv2_avg_path, output_path):
  # csv1 = pd.read_csv(csv1_path)
  csv1 = pd.read_csv(csv1_path, header=None)
  csv2_avg = pd.read_csv(csv2_avg_path)
  # Calculate cosine and L2 distances for each dimension
  results = []
  for _, row in csv1.iterrows():
    if 'base_image' in row:
      image = row['base_image']
      # csv1_probs 需要不包括最后一列 base_image
      csv1_probs = row.iloc[1:-1].values
    else:
      image = row[0]
      csv1_probs = row.iloc[1:].values
    # 是[2:]么？
    csv2_avg_probs = csv2_avg.loc[csv2_avg['base_image'] == image, csv2_avg.columns[1:]].values.flatten()

    if len(csv2_avg_probs) > 0:
      l2_dist, cosine_dist = calculate_distances(csv1_probs, csv2_avg_probs)
      cosines = [cosine_dist]
      l2s = [l2_dist]

      for k in [5, 20, 50]:
        logit1_top_k = retain_top_k(csv1_probs, k)
        logit2_top_k = retain_top_k(csv2_avg_probs, k)
        l2_top_k, cosine_top_k = calculate_distances(logit1_top_k, logit2_top_k)
        cosines.append(cosine_top_k)
        l2s.append(l2_top_k)

      results.append([image] + cosines + l2s)

  # Prepare column names for the output
  cosine_columns = [f'cosine_distance_{i}' for i in ["original", "top_5", "top_20", "top_50"]]
  l2_columns = [f'l2_distance_{i}' for i in ["original", "top_5", "top_20", "top_50"]]
  output_columns = ['image'] + cosine_columns + l2_columns

  # Save results to a new CSV
  results_df = pd.DataFrame(results, columns=output_columns)
  results_df.to_csv(output_path, index=False)

  print(f"Results saved to {output_path}")
  return results_df
````
